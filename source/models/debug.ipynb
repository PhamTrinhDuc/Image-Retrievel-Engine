```python
from resnet_extractor import ResNetExtractor
import torch
import numpy as np
from PIL import Image


device = "cuda" if torch.cuda.is_available() else "cpu"

extractor = ResNetExtractor(
    model_name="resnet34",
    device=device,
    batch_size=16,
    enable_mixed_precision=True if device == "cuda" else False
)
    
# 2. Model Information
print("\n2. Model Information")
model_info = extractor.get_model_info()
for key, value in model_info.items():
    print(f"   {key}: {value}")

# 3. Create dummy images for demonstration
print("\n3. Processing Demo Images")

# Create sample images
dummy_images = []
for i in range(5):
    # Create random RGB image
    img_array = np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8)
    img = Image.fromarray(img_array)
    dummy_images.append(img)

# 4. Single image processing
print("\n4. Single Image Feature Extraction")
single_features = extractor.get_embedding(dummy_images[0])
print(f"   Single image features shape: {single_features.shape}")
print(f"   Feature dimension: {extractor.feature_dim}")


# 5. Batch processing
print("\n5. Batch Feature Extraction")
batch_features = extractor.get_batch_embeddings(dummy_images, batch_size=3)
print(f"   Batch features shape: {batch_features.shape}")
print(f"   Processing {len(dummy_images)} images in batches of 3")
```
